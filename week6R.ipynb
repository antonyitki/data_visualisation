Activity 3: Basic Colour images in R

```r
# Data
x <- -10:10
y <- -10:10
z <- sqrt(outer(x ^ 2, y ^ 2, "+"))
image(x, y, z)
# You can also type, the following
# but the axes will be between 0 and 1
image(z)
```

```r
# Data
x <- -10:10
y <- -10:10
z <- sqrt(outer(x ^ 2, y ^ 2, "+"))

image(x, y, z, col = gray.colors(12))
```

```r
# Data
x <- -10:10
y <- -10:10
z <- sqrt(outer(x ^ 2, y ^ 2, "+"))

image(x, y, z, col = topo.colors(12))
```

```r
# Question 3) Adjust the number of colour values so the image will be smoothed.
# Data 
x <- -10:10
y <- -10:10
z <- sqrt(outer(x ^ 2, y ^ 2, "+"))

image(x, y, z)
contour(x, y, z, add = TRUE)
```
