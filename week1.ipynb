```python
#mean
mean1 = sum(dataset[0])/len(dataset[0])
print(mean1)
mean11 = np.mean(dataset[0]) #0 for row number
print(mean11)
```

```python
mean2 = np.mean(dataset[:, -1]) #: means all rows. -1 for last column
print(mean2)
```

```python
mean3 = np.mean(dataset[0:3, 0:3])
print(dataset[0:3, 0:3])
print(mean3)
```

```python
#median
np.median(dataset[-1])
```

```python
np.median(dataset[:,-3:]) #count from the end
```
