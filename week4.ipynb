**Activity 1: Understand Uncertainty**

```python
# Import the necessary modules and enable plotting within a jupyter notebook.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pandas.plotting import bootstrap_plot
```

```python
data = sns.load_dataset("penguins")
data = data.dropna()
```

```python
# make a histogram of bill length using Pandas hist() 
data.bill_length_mm.hist(bins=50)
```

```python
# make a histogram of bill length using Pandas hist() 
data.bill_length_mm.hist(bins=50).plot()
```

```python
# make a histogram of bill length using Pandas hist() 
data.bill_length_mm.hist(bins=50,figsize=(10,10))
```

```python
# Example: Sample 200 observation from our observed bill length with 
# replacement for 500 times and estimate the three statistics on each 
 
bootstrap_plot(data["bill_length_mm"],
               size=200,
               samples=500, 
               color="dodgerblue")

# bootstrap_plot() makes two types of plot for each of them and show how reliable our # estimate of mean/median/midrange is.
```

```python
bootstrap_plot(data["bill_length_mm"],
               size=300,
               samples=900, 
               color="red")
```
