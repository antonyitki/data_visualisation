```python
# Solution activity 1, step by step:
# Import the necessary modules and enable plotting within a jupyter notebook.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('/content/anage_data.csv')

# Preprocessing
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
#Question 1) Filter the data so you end up with samples containing a body mass and a maximum longevity.
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass])]
# Sort according to class
#Question 2) Select all samples of the class Aves and with a body mass smaller than 20,000
aves = data[data['Class'] == 'Aves']
aves = data[data[mass] < 20000]

# Create a figure with constrained layout. Create a GridSpec of size 4x4.
# Create a scatter plot of size 3x3 and marginal histograms of size 1x3 and 3x1.
#Add labels and a figure title.
#Question 3) Create figure
fig = plt.figure(figsize=(8, 8), dpi=150, constrained_layout=True)
# Create GridSpec
gs = fig.add_gridspec(4, 4)
#Question 4) Specify subplots
histx_ax = fig.add_subplot(gs[0, :-1])
histy_ax = fig.add_subplot(gs[1:, -1])
scatter_ax = fig.add_subplot(gs[1:, :-1])
# Create plots
scatter_ax.scatter(aves[mass], aves[longevity])
histx_ax.hist(aves[mass], bins=20, density=True)
histx_ax.set_xticks([])
histy_ax.hist(aves[longevity], bins=20, density=True, orientation='horizontal')
histy_ax.set_yticks([])
#Question 5) Add labels and title
plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')
fig.suptitle('Scatter plot with marginal histograms')
# Show plot
plt.show()

```
